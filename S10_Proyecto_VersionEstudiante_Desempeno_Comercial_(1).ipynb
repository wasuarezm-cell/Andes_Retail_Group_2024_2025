### Narrativa del Dashboard (Modelo SCQA)

#### 🖥️ Vista General (Overview)

**S (Situación):**  
Andes Retail Group generó USD 5.53 millones en ingresos totales con un margen de ganancia sólido del 35.02%. La operación está liderada por Perú y Chile, que en conjunto concentran cerca del 76% de los ingresos totales, sostenidos principalmente por los segmentos de clientes Premium y Estándar.

**C (Complicación):**  
A pesar del volumen general, se observa una fuerte dependencia estacional: las ventas alcanzan su pico a principios y a fin de año (diciembre y enero), pero caen significativamente durante el periodo de invierno. Además, Colombia presenta una importante brecha comercial, aportando solo el 24% de los ingresos. Los ingresos cayeron de USD 2.86 millones en 2024 a USD 2.67 millones en 2025, una baja de casi 7%.

**Q (Pregunta):**   
¿La caída de ingresos refleja un problema de rentabilidad, de concentración geográfica, o de otro factor estructural del negocio?

**A (Respuesta):**  
El margen se mantiene estable en 35%, por lo que la caída no está relacionada con rentabilidad. Perú y Chile concentran el 76% de los ingresos, dejando a Colombia como la geografía con mayor espacio de crecimiento. Dentro de los segmentos de cliente, Premium y Estándar sostienen el negocio casi por igual (47% y 44%), mientras Económico aporta apenas 9% — **no es prioritario resolver ahí**, pero tampoco tiene upside significativo sin una estrategia distinta. Es necesario investigar a nivel granular qué factores impulsan la caída de ventas a mitad de año para mitigar la estacionalidad.

---

#### 🔎 Vista Detalle

**S (Situación):**  
Al profundizar en el comportamiento estacional, confirmamos que las ventas sufren una caída crítica durante el invierno (USD 0.7 millones) en comparación con el verano (USD 2.2 millones). Al cruzar esta información con la rentabilidad, vemos que el margen se mantiene estable (~35%) sin importar el segmento de cliente.

**C (Complicación):**  
Dado que el margen es estable y la caída de ingresos en invierno es pareja en los tres países, confirmamos que el problema no radica en la estructura de costos, precios o en una geografía particular, sino en una pérdida transversal del volumen de demanda.

**Q (Pregunta):**  
¿La caída de ingresos está relacionada con un problema de rentabilidad o con un comportamiento estacional de la demanda?

**A (Respuesta):**  
La evidencia apunta a un fenómeno de estacionalidad más que a un problema de rentabilidad. El foco estratégico debe estar en incrementar el volumen de ventas durante el invierno mediante campañas estacionales, promociones y una mejor planificación comercial, ya que el margen permanece saludable independientemente del segmento.
